**Exercício 1:** crie três variáveis numéricas e exiba a soma delas.

````python
import math

def triangular(n):
    calculo = (8 * n) + 1
    raiz = math.sqrt(calculo)

    return raiz.is_integer()

n1 = int(input("Digite o primeiro numero: "))
n2 = int(input("Digite o segundo numero: "))
n3 = int(input("Digite o terceiro numero: "))

soma = n1 + n2 + n3

if triangular(soma):
    print("E triangular")
else:
    print("O numero nao e triangular.")
````

**Exercício 14**

Verifique se uma palavra é palíndromo.

````python
def palindromo(texto):
    texto_limpo = texto.replace(" ", "").lower()
    return texto_limpo == texto_limpo[::-1]

texto = input("Digite uma palavra qualquer: ")
if palindromo(texto):
    print("A palavra e um palindromo.")
else:
    print("A palavra nao e um palindromo.")
````

**Seção 11 — Raise**

**Exercício 18**

Crie uma função chamada `validar_senha(senha)` que deve verificar:

* A senha deve ter pelo menos 8 caracteres
* Deve conter pelo menos um número
* Deve conter pelo menos uma letra maiúscula

Se qualquer uma dessas regras não for cumprida, a função deve lançar uma exceção personalizada chamada:

```
class SenhaInvalidaError(Exception):
    pass
```



A função deve retornar `"Senha válida!"` se tudo estiver correto.

````python
class SenhaInvalidaError(Exception):
    pass

def validar_senha(senha):
    if len(senha) < 8:
        raise SenhaInvalidaError("A senha deve ter pelo menos 8 caracteres.")
    if not any(char.isdigit() for char in senha):
        raise SenhaInvalidaError("A senha deve conter pelo menos um numero.")
    if not any(char.isupper() for char in senha):
        raise SenhaInvalidaError("A senha deve conter pelo menos uma letra maiuscula.")
    return "Senha valida"
try:
    print(validar_senha("Senha"))
except SenhaInvalidaError as e:
    print(f"Erro de validacao: {e}")
````

**Exercício 19**

Crie um programa que:

1. Peça nome e idade de 5 pessoas

2. Armazene em uma lista de dicionários

3. Informe quem é a pessoa mais velha

````python
lista = []
for i in range(5):
    nome = input("Digite o seu nome: ")
    idade = int(input("Digite a sua idade: "))
    pessoa = {"nome": nome, "idade": idade}
    lista.append(pessoa)

mais_velha = lista[0]
for pessoa in lista:
    if pessoa['idade'] > mais_velha['idade']:
        mais_velha = pessoa
print(f"\n A mais velha {mais_velha['nome']} com {mais_velha['idade']} anos.")
````
